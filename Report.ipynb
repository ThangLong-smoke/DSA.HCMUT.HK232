Hệ thống mô phỏng quy trình vận hành tại của hàng mắt kính, trong đó khách hàng đến để **đăng ký sổ khám mắt**, **kiểm tra mắt**, **tư vấn chọn gọng kính**, **cắt kính**, **lắp kính** và **thanh toán**.

Mỗi bước dịch vụ sẽ có thời gian xử lý cụ thể và thời gian chờ, ảnh hưởng đến tổng thời gian phục vụ khách hàng.




## Hành vi của hệ thống
Khi khách hàng đến cửa hàng mắt kính sẽ có 2 trường hợp:
- Khách hàng đi kiểm tra mắt và cắt kính.
- Khách hàng đi thay gọng kính.

Đối với trường hợp khi khách hàng đi kiểm tra mắt và cắt kính thì bước đầu tiên khách hàng sẽ tiến hành đăng ký sổ kiểm tra mắt (register check-up book). Hệ thống sẽ có 2 quầy đăng ký sổ riêng biệt, mỗi quầy sẽ có 1 hàng đợi tương ứng. Mỗi quầy để đăng ký sổ được mô hình bằng một hệ thống hàng đợi M/M/1. Đặc điểm của hệ thống hàng đợi đăng ký sổ như sau:
- Thời gian giữa 2 khách hàng khi đến đăng ký và thời gian xử lý của server đều có phân phối Poisson.
- Một server.
- Không có giới hạn về số lượng người trong hàng chờ.
- Phục vụ theo nguyên tắc First-come-first-served.

Việc lựa chọn quầy đăng ký của khách hàng không bị ràng buộc bởi bất kỳ yếu tố nào của cửa hàng. Đặt xác suất khách hàng chọn quầy đăng ký 1 là \( p_{01} \) và quầy đăng ký 2 là \( p_{02} \). Xác suất để khách hàng chọn quầy đăng ký 1 là \( p_{01} \) và xác suất để khách hàng chọn quầy đăng ký 2 là \( p_{02} \).

Ngoài ra, đối với những khách hàng đã đăng ký sổ khám mắt từ trước có thể bỏ qua bước đăng ký sổ của cửa hàng và tiến đến hàng đợi kiểm tra mắt. Xác suất một khách hàng đã đăng ký sổ từ trước là \( p_{03} \).

Đối với những khách hàng chỉ có nhu cầu đến lắp gọng kính thì có thể tiến thẳng đến quầy tư vấn gọng kính. Xác suất để khách hàng đến cửa hàng chỉ để thay gọng kính là \( p_{04} \). Tổng xác suất của \( p_{01}, p_{02}, p_{03}, p_{04} \) bằng 1.

Trong bước đăng ký sổ, sẽ có trường hợp khách hàng không mang đủ giấy tờ tùy thân để đăng ký sổ khám, do đó khách hàng không được tiếp tục sử dụng dịch vụ khám và kiểm tra mắt của cửa hàng. Dẫn tới, khách phải rời của hàng đối với trường hợp này. Xác suất để khách đăng ký sổ và không mang đủ giấy tờ tùy thân là \( p_{1\_11} \) đối với quầy đăng ký 1 và \( p_{2\_11} \) đối với quầy đăng ký 2.

Sau khi hoàn thành đăng ký sổ, khách hàng sẽ tiến hành đo mắt và khám mắt. Tất cả khách hàng sau khi đăng ký sẽ xếp thành một hàng duy nhất để được vào thăm khám. Hệ thống đo và kiểm tra mắt được mô hình bằng một hệ thống hàng đợi M/M/c có đặc điểm như sau:
- Thời gian giữa 2 khách hàng khi đến kiểm tra mắt và thời gian xử lý của server có phân phối Poisson.
- Nhiều server.
- Không có giới hạn về số lượng người trong hàng chờ.
- Phục vụ theo nguyên tắc First-come-first-served.

Sau giai đoạn kiểm tra và khám mắt, khách hàng có thể lựa chọn cắt kính mới hoặc giữ kính cũ và tiến hành thanh toán.

Đối với trường hợp khách tiến hành xếp hàng để thanh toán ngay sau giai đoạn kiểm tra mắt có xác suất là \( p_{5} \).

Đối với trường hợp khách muốn cắt kính mới, cửa hàng sẽ có 3 quầy để cắt kính riêng biệt. Mỗi quầy cắt kính sẽ được mô hình hóa bằng mô hình M/M/1. Xác suất để khách hàng vào quầy 1, 2 và 3 lần lượt là \( p_{6}, p_{7}, p_{8} \). Đặc điểm của từng quầy cắt kính:
- Thời gian giữa 2 khách hàng khi đến kiểm tra mắt và thời gian xử lý của server có phân phối Poisson.
- Một server.
- Không có giới hạn về số lượng người trong hàng chờ.
- Phục vụ theo nguyên tắc First-come-first-served.

Sau khi hoàn thành bước cắt kính, đối với quầy cắt kính 1, khách hàng sẽ tiếp tục chọn để xếp vào một quầy lắp kính. Trong cửa hàng sẽ có 3 quầy lắp kính cho khách hàng lựa chọn tự do. Xác suất để người dùng vào 1 trong 3 quầy tương ứng với \( p_{9}, p_{10}, p_{11} \). Tương ứng với khách ra khỏi quầy cắt kính 2 và 3 lần lượt là \( p_{12}, p_{13}, p_{14} \) và \( p_{15}, p_{16}, p_{17} \). Mỗi quầy lắp kính sẽ tuân theo mô hình hàng chờ M/M/1 có đặc điểm sau:
- Thời gian giữa 2 khách hàng khi đến kiểm tra mắt và thời gian xử lý của server có phân phối Poisson.
- Một server.
- Không có giới hạn về số lượng người trong hàng chờ.
- Phục vụ theo nguyên tắc First-come-first-served.

Sau khi lắp kính, khách hàng sẽ phải tiến đến xếp hàng vào quầy thanh toán. Trong cửa hàng sẽ chỉ có 1 quầy thanh toán và tuân theo mô hình hàng chờ M/M/1 có đặc điểm sau:
- Thời gian giữa 2 khách hàng khi đến kiểm tra mắt và thời gian xử lý của server có phân phối Poisson.
- Một server.
- Không có giới hạn về số lượng người trong hàng chờ.
- Phục vụ theo nguyên tắc First-come-first-served.

Sau bước thanh toán, khách hàng sẽ được tính là thanh toán thành công và hoàn thành toàn bộ quá trình.





## Performance Evaluation Step

### State Goals and Define the System
- **State goals**
  - Ước lượng sự ảnh hưởng của lượng khách hàng đến thời gian phục vụ của các bộ phận trong phòng khám mắt.
  - Kiểm tra tính ổn định của hệ thống đối với các lượng khách hàng khác nhau.
  - Đánh giá được bộ phận là nút nghẽn cổ chai (bottleneck) của cả hệ thống.
- **System boundary**
  - Lượng khách hàng đến phòng khám trong 1 giờ.
  - Thời gian phục vụ đăng ký sổ khám cho mỗi khách hàng.
  - Thời gian chờ khi đăng ký sổ của mỗi khách hàng.
  - Thời gian kiểm tra mắt cho mỗi khách hàng.
  - Thời gian chờ đến lượt kiểm tra mắt của mỗi khách hàng.
  - Thời gian cắt kính cho mỗi khách hàng.
  - Thời gian chờ đến lượt cắt kính.
  - Thời gian lắp kính của mỗi khách hàng.
  - Thời gian chờ đến lượt để lắp kính của mỗi khách hàng.
  - Thời gian phục vụ thanh toán của mỗi khách hàng.
  - Thời gian chờ đến lượt để thanh toán của mỗi khách hàng.

### Service and Outcomes:
- **Service: Đăng ký sổ khám bệnh**
  - Khách hàng đến phòng khám, cung cấp thông tin cá nhân và đăng kí hoặc xác nhận lịch hẹn đã đặt trước. Nhân viên kiểm tra và lưu thông tin trong hệ thống để sắp xếp vị trí cho khách.
  - **Outcome:**
    - Đăng ký thành công và khách hàng được sắp xếp vào hàng chờ.
    - Đăng ký thành công và khách hàng được xác nhận lịch hẹn.
    - Đăng ký không thành công vì khách hàng không mang đầy đủ giấy tờ, do đó khách hàng đi về.
    - Không có lịch trống, khách hàng đi về.
    
- **Service: Kiểm tra mắt**
  - Khách hàng sau khi đăng ký sẽ được hướng dẫn vào phòng khám mắt và trải qua quá trình kiểm tra thị lực và khám mắt.
  - **Outcome:**
    - Khách hàng được kiểm tra thành công.
    - Khách hàng bỏ qua kiểm tra mắt và không tiếp tục dịch vụ cắt kính.

- **Service: Cắt kính**
  - Khách hàng sau khi kiểm tra mắt sẽ được tư vấn về kính thuốc và được cắt kính theo nhu cầu.
  - **Outcome:**
    - Khách hàng cắt kính thành công.
    - Khách hàng không cắt kính mà chỉ lấy kính cũ.

- **Service: Lắp kính**
  - Khách hàng đến quầy lắp kính để lắp kính mới đã cắt.
  - **Outcome:**
    - Khách hàng lắp kính thành công.
    - Khách hàng không lắp kính.

- **Service: Thanh toán**
  - Khách hàng sau khi lắp kính sẽ tiến hành thanh toán.
  - **Outcome:**
    - Thanh toán thành công.
    - Thanh toán không thành công do lỗi hệ thống.
    
### Performance Metrics:
- **Hoàn thành quy trình:** Số lượng khách hàng hoàn thành quy trình trong 1 giờ.
- **Thời gian phục vụ trung bình:**
  - Thời gian phục vụ trung bình cho từng bước trong quy trình (đăng ký, kiểm tra mắt, cắt kính, lắp kính, thanh toán).
  - Thời gian chờ đến lượt cho từng bước trong quy trình.
  
- **Done incorrectly:** Không hoàn thành quy trình
  - Số lượng khách hàng không mang đủ giấy tờ để đăng ký sổ khám.
  - Tỷ lệ khách hàng không xác nhận được lịch hẹn hoặc không còn lịch trống.
  - Tỷ lệ khách hàng bỏ qua bước kiểm tra mắt hoặc không tiếp tục dịch vụ cắt kính sau khi khám.

### Data Collection:
Để đánh giá hiệu suất của hệ thống, việc thu thập dữ liệu là rất quan trọng. Các loại dữ liệu cần thu thập bao gồm:
- Số lượng khách hàng đến cửa hàng theo giờ.
- Thời gian thực tế cho từng giai đoạn dịch vụ (đăng ký, kiểm tra mắt, cắt kính, lắp kính, thanh toán).
- Số lượng khách hàng trong hàng đợi tại mỗi giai đoạn dịch vụ.
- Tỷ lệ khách hàng từ chối dịch vụ do không đủ giấy tờ.

Các dữ liệu này có thể được thu thập thông qua:
- Hệ thống quản lý thông tin khách hàng (CRM) để lưu trữ và phân tích thông tin khách hàng.
- Bảng điều khiển (dashboard) theo dõi thời gian thực cho từng giai đoạn dịch vụ.
- Khảo sát khách hàng để thu thập phản hồi về dịch vụ.

### Analysis:
Khi đã thu thập đủ dữ liệu, các phân tích cần được thực hiện để xác định hiệu suất của hệ thống:
- Phân tích thống kê để xác định thời gian trung bình cho từng giai đoạn và so sánh với mục tiêu đã đề ra.
- Xác định các nút thắt cổ chai trong quy trình phục vụ và tìm kiếm giải pháp để cải thiện hiệu suất.
- Đánh giá tỷ lệ hoàn thành quy trình dịch vụ để xác định những nguyên nhân dẫn đến việc khách hàng không hoàn thành dịch vụ.

Kết quả của các phân tích này sẽ giúp cửa hàng điều chỉnh quy trình làm việc, cải thiện dịch vụ khách hàng và tối ưu hóa hiệu suất hoạt động.
